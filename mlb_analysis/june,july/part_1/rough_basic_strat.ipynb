```python
import pandas as pd
import datetime
from datetime import datetime as dt
from datetime import date
import matplotlib.pyplot as plt

df = pd.read_csv("data/mlb_2022_season_data.csv", index_col=0)

# win / loss = df1
sdate = date(2022, 3, 30)
edate = date(2022, 7, 10)
df1 = pd.DataFrame()

while True:
    if (sdate > edate): 
        break
    sDate = sdate.strftime('%Y-%m-%d')
    df1[f'+/- ({sDate})'] = df[f'+/- ({sDate})']
    sdate = sdate + datetime.timedelta(days = 1)

df1.index.name = 'team'


# points scored = df3
sdate = date(2022, 3, 30)
edate = date(2022, 7, 10)
df3 = pd.DataFrame()

while True:
    if (sdate > edate): 
        break
    sDate = sdate.strftime('%Y-%m-%d')
    df3[f'points scored ({sDate})'] = df[f'points scored ({sDate})']
    sdate = sdate + datetime.timedelta(days = 1)

df3.index.name = 'team'





# win / loss cummulative sum = df2
sdate = date(2022, 3, 30)
edate = date(2022, 7, 10)
df2 = pd.DataFrame()
df2 = df1.cumsum(axis=1)

while True:
    if (sdate > edate): 
        break
    sDate = sdate.strftime('%Y-%m-%d')
    df[f'+/- ({sDate})'] = df2[f'+/- ({sDate})']
    sdate = sdate + datetime.timedelta(days = 1)



# points scored cummulative sum = df4
sdate = date(2022, 3, 30)
edate = date(2022, 7, 10)
df4 = pd.DataFrame()
df4 = df3.cumsum(axis=1)

while True:
    if (sdate > edate): 
        break
    sDate = sdate.strftime('%Y-%m-%d')
    df[f'points scored ({sDate})'] = df4[f'points scored ({sDate})']
    sdate = sdate + datetime.timedelta(days = 1)

df
```
